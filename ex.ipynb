```python
# Lista de itens fictícios
nomes = [
    'Pizza Margherita', 'Hambúrguer', 'Salada Caesar', 'Espaguete à Carbonara', 'Tiramisu',
    'Lasagna', 'Calzone', 'Ravioli', 'Bruschetta', 'Risotto', 
    'Sopa de Tomate', 'Bife à Milanesa', 'Frango Grelhado', 'Canelone', 'Gnocchi',
    'Mousse de Chocolate', 'Panacotta', 'Creme Brûlée', 'Brownie', 'Cheesecake',
    'Batata Frita', 'Onion Rings', 'Almôndegas', 'Pão de Alho', 'Ceviche',
    'Fraldinha', 'Costela', 'Salmão Grelhado', 'Camarão ao Alho', 'Filet Mignon',
    'Pizza Pepperoni', 'Spaghetti ao Pesto', 'Salada Caprese', 'Lasanha Vegetariana', 'Pavê',
    'Pão de Queijo', 'Salada de Frutas', 'Tacos', 'Quesadilla', 'Churros'
]

# Criando dados aleatórios para preços, custos e quantidades vendidas
precos = np.random.uniform(10.0, 35.0, size=len(nomes)).round(2)
custos = np.random.uniform(5.0, 20.0, size=len(nomes)).round(2)
quantidades_vendidas = np.random.randint(50, 200, size=len(nomes))
```

```python
# Criando o DataFrame
data = {
    'Nome': nomes,
    'Categoria': np.random.choice(['Prato Principal', 'Entrada', 'Sobremesa'], size=len(nomes)),
    'Preço': precos,
    'Custo': custos,
    'Quantidade Vendida': quantidades_vendidas
}

df_pizza0 = pd.DataFrame(data)
# Calculando o lucro
df_pizza0['Lucro'] = (df['Preço'] - df['Custo']) * df['Quantidade Vendida']
```

```python
df_pizza0
```

```python
#separando por categoria
df_PP = df_pizza0.loc[df_pizza0['Categoria'] =='Prato Principal']
df_sobremesa = df_pizza0.loc[df_pizza0['Categoria'] =='Sobremesa']
df_entrada = df_pizza0.loc[df_pizza0['Categoria'] =='Entrada']
```

```python
#arrumando valores
df_PP_sort = df_PP.sort_values(by='Lucro',ascending=False)
df_sobremesa_sort = df_sobremesa.sort_values(by='Lucro',ascending=False)
df_entrada_sort = df_entrada.sort_values(by='Lucro',ascending=False)
```
